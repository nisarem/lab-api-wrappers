```python
import spotipy
import pandas as pd

from spotipy.oauth2 import SpotifyClientCredentials
from pandas import json_normalize
```

```python
string = open("access.txt","r").read()
```

```python
access={}
for line in string.split('\n'):
    if len(line) > 0:
        access[line.split(':')[0]]=line.split(':')[1].strip()
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=access['clientid'],
                                                           client_secret=access['clientsecret']))
```

```python
from random import randint
from time import sleep

def get_playlist(user, num_playlists):
    playlist = []
    total_playlists = 0
    results = sp.user_playlists(user)  
    playlist.extend(results['items'])
    total_playlists += len(results['items'])
    while total_playlists < num_playlists and results['next']: 
        results = sp.next(results) 
        playlist.extend(results['items'])  
        total_playlists += len(results['items'])
        sleep(randint(1, 3))
        playlists = pd.json_normalize(playlist[:num_playlists])
    return playlists
```

```python
playlists = get_playlist('Spotify',150)
```

```python
spotify_playlists = pd.DataFrame(columns=['href', 'id', 'name','public','uri','tracks.total'])
```

```python
spotify_playlists_data = []
    
for index, row in playlists.iterrows():
    names = row['name']
    track_count = row['tracks.total']
    ids = row['id']
    uri = row['uri']
    href = row['href']
    public = row['public']
    owner= row['owner.id']
    
    playlist_info = {
        'name': names,
        'track_count': track_count,
        'id': ids,
        'uri': uri,
        'href': href,
        'public': public,
        'owner_id': owner
    }
    
    spotify_playlists_data.append(playlist_info)

spotify_playlists = pd.DataFrame(spotify_playlists_data)    
spotify_playlists     
```

```python
def get_tracks(user, playlist_id):
    results = sp.user_playlist_tracks(user, playlist_id)
    track= results['items']
    while results['next']:
        results = sp.next(results)
        track.extend(results['items'])
    tracks = pd.json_normalize(track)
    return tracks
```

```python
def get_playlist_tracks(user, playlist_id):
    results = sp.user_playlist_tracks(user,playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) 
    return tracks
```

```python
def get_all_tracks(playlists):
    all_tracks = []
    for _, row in playlists.iterrows():
        tracks = get_playlist_tracks(row['owner_id'], row['id'])
        all_tracks.extend(tracks)
        sleep(randint(1, 3))  
    return all_tracks
```

```python
all_tracks = get_all_tracks(spotify_playlists)
all_tracks_df = pd.DataFrame(all_tracks)
```
