```python
from random import randint
from time import sleep

def get_playlist(user, num_playlists):
    playlist = []
    total_playlists = 0
    results = sp.user_playlists(user)  
    playlist.extend(results['items'])
    total_playlists += len(results['items'])
    while total_playlists < num_playlists and results['next']: 
        results = sp.next(results) 
        playlist.extend(results['items'])  
        total_playlists += len(results['items'])
        sleep(randint(1, 3))
        playlists = pd.json_normalize(playlist[:num_playlists])
    return playlists
```

```python
spotify_playlists = pd.DataFrame(columns= ['href', 'id', 'name', 'public', 
                                           'uri', 'owner.id','tracks.total'])
```

```python
spotify_playlists_data = []
    
for index, row in playlists.iterrows():
    names = row['name']
    track_count = row['tracks.total']
    ids = row['id']
    uri = row['uri']
    href = row['href']
    public = row['public']
    owner= row['owner.id']
    
    playlist_info = {
        'name': names,
        'track_count': track_count,
        'id': ids,
        'uri': uri,
        'href': href,
        'public': public,
        'owner_id': owner
    }
    
    spotify_playlists_data.append(playlist_info)

spotify_playlists = pd.DataFrame(spotify_playlists_data)    
spotify_playlists     
```

```python
def get_tracks(user, playlist_id):
    results = sp.user_playlist_tracks(user, playlist_id)
    track= results['items']
    while results['next']:
        results = sp.next(results)
        track.extend(results['items'])
    tracks = pd.json_normalize(track)
    return tracks
```

```python
result_list = []
for index, row in spotify_playlists.iterrows():
    user_id = row['owner_id']
    playlist_id = row['id']
    tracks = get_tracks(user_id, playlist_id)
    result_list.append(tracks)

all_tracks = pd.concat(result_list, ignore_index=True)
```

```python
import pandas as pd

artists_df = pd.DataFrame(columns=['href', 'id', 'name', 'type', 'uri', 'external_urls.spotify','song_id', 'song_name', 'popularity'])

for i, artists_info in enumerate(all_tracks['track.artists']):
    if isinstance(artists_info, list):  # Check if artists_info is a list
        for artist in artists_info:
            if pd.notna(artist):  # Check if artist is not NaN
                artist_df = pd.json_normalize(artist)
                song_id = all_tracks.iloc[i]['track.id']
                song_name = all_tracks.iloc[i]['track.name']
                popularity = all_tracks.iloc[i]['track.popularity']
                
                artist_df['song_id'] = song_id
                artist_df['song_name'] = song_name
                artist_df['popularity'] = popularity
                
                artists_df = pd.concat([artists_df, artist_df], axis=0, ignore_index=True)

artists_df.reset_index(drop=True, inplace=True)
```

```python
artists_df
```

```python
df = artists_df[['song_name', 'name', 'song_id', 'popularity']]
df
```

```python
chunks = [(i, i+100) for i in range(0, len(df), 100)]
chunks
audio_features_list = []
for chunk in chunks:
    id_list100 = df['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
audio_features_df.drop_duplicates(inplace=True) # duplicates because some songs have more artists
audio_features_df = audio_features_df.reset_index(drop=True)
audio_features_df
```

```python
df_w_audio_ft = pd.merge(left=df,
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```
